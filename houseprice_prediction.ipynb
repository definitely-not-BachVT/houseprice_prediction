```python
# Features: sqft_living, sqft_lot, bedrooms,
# bathrooms, floors, waterfront, view, condition, yr_built

numeric_cols = df.select_dtypes(include = ['int64', 'float64'])
numeric_cols.head()
```

```python
positive_price = len(df[df['price'] > 0])
positive_price
# 4091 < 4140
non_negative_bedrooms = len(df[df['bedrooms'] >= 0])
non_negative_bedrooms
# 4140
non_negative_bathrooms = len(df[df['bathrooms'] >= 0])
non_negative_bathrooms
# 4140
```

```python
numeric_cols = numeric_cols[numeric_cols >= 0]
numeric_cols = numeric_cols[numeric_cols['price'] > 0]
numeric_cols.head()
```

```python
df = numeric_cols
# yr_renovated nếu không có thì ghi là 0, đang có chênh lệch rất lớn
# => Đổi thành có được tu sửa không nhé
df.loc[df['yr_renovated'] > 0, 'yr_renovated'] = 1
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns = ['price'])
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 36)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha = 1.0, solver = 'auto', random_state = 36)
model.fit(scaled_X_train, y_train)

y_pred = model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(mae)
print(mse)
print(r2)
```
